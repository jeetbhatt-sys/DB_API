# DataBase Creation API

URL : http://127.0.0.1:5000/db_operations

parameters : 
{
    "host" : "localhost", 
    "user" : "root",
    "passwd" : "mysql",
    "dbName" : "Employee"
}

```python
from flask import Flask, render_template, request, jsonify
import mysql.connector as connection

app = Flask(__name__)

def CreateDB(host,user,passwd,dbName):

    try:
        mydb = connection.connect(host=host, user=user, passwd=passwd, use_pure=True)
        print(mydb.is_connected())
        query = "Create database " + dbName +';'
        cursor = mydb.cursor()  # create a cursor to execute queries
        cursor.execute(query)
        print("Database Created!!")
        mydb.close()
        return 1
    except Exception as e:
        mydb.close()
        return -1

@app.route('/db_operations', methods=['POST'])
def db_operation():
    if request.method == 'POST':
        host = request.json['host']
        user = request.json['user']
        passwd = request.json['passwd']
        dbName = request.json['dbName']
        res = CreateDB(host, user, passwd,dbName)
        if res == -1:
            return '''<h1> Failed to create the Data Base.</h1> '''
        else:
            return '''<h1> Data Base Created successfully.</h1> '''


if __name__ == '__main__':
    app.run()
```

# CreateTable

URL : http://127.0.0.1:5000/CreateTable

parameters : 
{
    "host" : "localhost", 
    "user" : "root",
    "passwd" : "mysql",
    "dbName" : "Employee",
    "query" : "CREATE TABLE employeeDetails (Empid INT(10) AUTO_INCREMENT PRIMARY KEY,FirstName VARCHAR(60),LastName VARCHAR(60), JoiningDate DATE,Class Varchar(20), JobGrad Varchar(10))"
}

```python
from flask import Flask, render_template, request, jsonify
import mysql.connector as connection

app = Flask(__name__)

def CreateTable(host,user,passwd,dbName,query):
    import mysql.connector as connection

    try:
        mydb = connection.connect(host=host, user=user,database = dbName, passwd=passwd, use_pure=True)
        # check if the connection is established
        query = query
        cursor = mydb.cursor() #create a cursor to execute queries
        cursor.execute(query)
        mydb.close()
        return 1
    except Exception as e:
        mydb.close()
        return -1    

@app.route('/CreateTable', methods=['POST'])
def db_operation():
    if request.method == 'POST':
        host = request.json['host']
        user = request.json['user']
        passwd = request.json['passwd']
        dbName = request.json['dbName']
        query = request.json['query']
        res = CreateTable(host, user, passwd,dbName,query)
        if res == -1:
            return '''<h1> Failed to create the Table.</h1> '''
        else:
            return '''<h1> Table Created successfully.</h1> '''


if __name__ == '__main__':
    app.run()
```

```python
Bulk data upload

URL : http://127.0.0.1:5000/BulkDataUploadToTable

Parameters : 
{
    "host" : "localhost", 
    "user" : "root",
    "passwd" : "mysql",
    "dbName" : "Employee",
    "tableName" : "employeeDetails",
    "csvFile" : "employee.csv"
}
```

```python
from flask import Flask, render_template, request, jsonify
import mysql.connector as connection
import pandas as pandas
import csv
app = Flask(__name__)

def BulkDataUploadToTable(host,user,passwd,dbName,tableName,csvFile):
    try:
        mydb = connection.connect(host=host, user=user,database = dbName, passwd=passwd, use_pure=True)
        cursor = mydb.cursor() 
        with open(csvFile, "r") as f:
            next(f)
            emp_data = csv.reader(f, delimiter="\n")
            for line in enumerate(emp_data):
                for list_ in (line[1]):
                    print(list_)
                    cursor.execute('INSERT INTO ' + tableName + ' values ({values})'.format(values=(list_)))
        mydb.commit()
        cursor.close()
        mydb.close()
        return 1
    except Exception as e:
        mydb.close()
        return -1
@app.route('/BulkDataUploadToTable', methods=['POST'])
def db_operation():
    if request.method == 'POST':
        host = request.json['host']
        user = request.json['user']
        passwd = request.json['passwd']
        dbName = request.json['dbName']
        tableName = request.json['tableName']
        csvFile = request.json['csvFile']
        res = BulkDataUploadToTable(host, user, passwd,dbName,tableName,csvFile)
        if res == -1:
            return '''<h1> Failed to Upload Bulk Data the Table.</h1> '''
        else:
            return '''<h1>Bulk Upload Data successfully.</h1> '''


if __name__ == '__main__':
    app.run()
```

```python
Data Fetch

URL : http://127.0.0.1:5000/GetDataFromTable

Parameter :
{
    "host" : "localhost", 
    "user" : "root",
    "passwd" : "mysql",
    "dbName" : "Employee",
    "tableName" : "employeeDetails"
}
```

```python
from flask import Flask, render_template, request, jsonify
import mysql.connector as connection
import pandas as pd
app = Flask(__name__)

def GetDataFromTable(host,user,passwd,dbName,tableName):
    try:
        mydb = connection.connect(host=host, user=user,database = dbName, passwd=passwd, use_pure=True)
        query = "Select * from "+ tableName +";"
        result_dataFrame = pd.read_sql(query,mydb)
        mydb.close()
        return result_dataFrame
    except Exception as e:
        return "Can't able to fetch the data."
    
@app.route('/GetDataFromTable', methods=['POST'])
def db_operation():
    if request.method == 'POST':
        host = request.json['host']
        user = request.json['user']
        passwd = request.json['passwd']
        dbName = request.json['dbName']
        tableName = request.json['tableName']
        res = GetDataFromTable(host, user, passwd,dbName,tableName)
        return '''{}'''.format(res)

if __name__ == '__main__':
    app.run()
```
